```python
import uxarray as ux
import cartopy.crs as ccrs 
```

```python
dir_data = "/glade/derecho/scratch/duda/mpas_tutorial/240-48km_variable/"

flnm_grid = dir_data + "SouthAmerica.init.nc"
flnm_data = flnm_grid 

uxds_mpas = ux.open_dataset(flnm_grid, flnm_data)

uxgrid = uxds_mpas.uxgrid 
```

```python
proj = ccrs.Orthographic(central_latitude=20, central_longitude=80)

plot_mesh = uxgrid.plot.edges(projection=proj, features=["states", "coastline"],
                  height=500, width=750,
                  linewidth = .2,
                  color="black", 
                  periodic_elements="split",
                  title="Grid Edge Plot")

plot_mesh 
```

#### Save plot

```python
import holoviews as hv
hv.save(plot_mesh , 'regional_mesh_my', fmt='pdf') 
```
