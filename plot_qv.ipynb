#### Plotting MPAS unstructured triangular grid through the UxDataArray.plot() accessor.  
 - Switching between Bokeh and Matplotlib backends for different rendering capabilities
   - Bokeh serves as the default renderer. 
   - Switch to Matplotlib by specifying the backend parameter
 - Periodic Data: periodic_elements='split': Periodic polygons are split along the antimeridian

```python
import uxarray as ux
```

```python
dir_data = '/glade/derecho/scratch/duda/mpas_tutorial/240-48km_variable/'
flnm_grid = dir_data + 'SouthAmerica.static.nc' 
flnm_data = dir_data + 'history.2014-09-11_00.00.00.nc'
ds_i = ux.open_dataset(flnm_grid, flnm_data)
qv = ds_i.qv[0,:,0]
```

#### UXarray seamless integrats with hvPlot, tailors plot and makes it interactive

```python
qv.plot(backend='bokeh', cmap='Grays', clim = (0, 0.02), features=["states", "coastline"],
           periodic_elements="split", title=qv.long_name)
```

#### Switch to Matplotlib by specifying the backend parameter

```python
qv.plot.polygons(backend='matplotlib', pixel_ratio = 4.0, features=["states", "coastline"],
         width=1000, height=500, cmap='Gray', clim = (0, qv.max()),
         periodic_elements="split", title= 'Lowest model level ' + qv.long_name + 'kg/kg')
```
